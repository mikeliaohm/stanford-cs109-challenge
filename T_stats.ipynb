### Estimate Race Window

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to read jitter values from the file
def collect_race_windows_values(file_path):
    race_window_values = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            try:
                race_window = float(line)
                race_window_values.append(race_window)
            except ValueError:
                # Skip lines that do not match the expected format
                continue
    return race_window_values

def print_stats(race_window_values):
     mean = np.mean(race_window_values)
     std = np.std(race_window_values)
     print(f"min {min(race_window_values):.4f}\tmax {max(race_window_values):.4f}")
     print(f"mean {mean:.4f}\tstd dev {std:.4f}")
     print(f"Number of samples {len(race_window_values)}")

def display_histogram(race_window_values):
	plt.hist(race_window_values, bins=50, density=True)
	plt.title(f"Race Window Histogram")
	plt.xlabel('Race Window (ms)')
	plt.ylabel('Frequency')
	plt.show()
```

```python
values = collect_race_windows_values("./Reports/race_window_reported.txt")
print_stats(values)
display_histogram(values)
```
